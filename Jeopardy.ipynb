```python
import pandas as pd
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Assuming your file is in 'MyDrive', adjust the path accordingly
file_path = '/content/drive/MyDrive/jeopardy.csv'  # Update with the correct path

# Read the CSV file using the full file path
df_raw = pd.read_csv(file_path)

# ... rest of your code ...
df_raw.rename(columns=lambda x: x.replace(" ", ""), inplace=True)
print(df_raw.columns)
print(df_raw)

def filter_data_frame_by_words(dataframe, column_name, words_to_check):
  words_to_check = [word.lower() for word in words_to_check]
  filtered_df = dataframe[dataframe[column_name].apply(lambda x: all(word in x.lower() for word in words_to_check))]
  return filtered_df

#call the function
word_list= ['The', 'of']
filtered_dataframe = filter_data_frame_by_words(df_raw, 'Question', word_list)
print(filtered_dataframe)

#clean the rows
df_raw['Value'] = df_raw['Value'].str.strip("'")
df_raw['Value'] = df_raw['Value'].str.replace(',' , '')
df_raw['Value'] = df_raw['Value'].str.replace('$' , '')
print(df_raw['Value'])
df_raw['Value'] = pd.to_numeric(df_raw['Value'], errors='coerce')

#organize data
av_val = df_raw.groupby('Category')['Value'].mean().reset_index()

#inspect data
print(av_val)
print(df_raw.columns)

#define the function
def count_unique(dataframe, question_column, answer_column):
  new_values = dataframe.groupby(question_column)[answer_column].nunique()
  return new_values

#call the function
count_answers = count_unique(df_raw, 'Question', 'Answer')
print(count_answers)




```

```python
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Assuming your file is in 'MyDrive', adjust the path accordingly
file_path = '/content/drive/MyDrive/jeopardy.csv'  # Update with the correct path

# Read the CSV file using the full file path
jeopardy_df = pd.read_csv(file_path)

# Changing 'category' to 'Category' to match the actual column name
plt.figure(figsize=(12, 6))
category_counts = jeopardy_df['Category'].value_counts().head(10) # Changed to 'Category'
sns.barplot(x=category_counts.values, y=category_counts.index, palette='coolwarm')
plt.xlabel('Number of Questions')
plt.ylabel('Category')
plt.title('Top 10 Most Common Jeopardy Categories')
plt.show()


### üìå ** Difficulty Distribution Histogram**

plt.figure(figsize=(10, 5))
# Changing 'value' to 'Value' to match the actual column name
sns.histplot(jeopardy_df['Value'].dropna(), bins=20, kde=True, color='blue') # Changed to 'Value'
plt.xlabel('Question Value ($)')
plt.ylabel('Frequency')
plt.title('Distribution of Jeopardy Question Values')
plt.show()


### üìå **Time Series Analysis**

# Assuming 'air_date' is a column containing dates, you might need to convert it to datetime
jeopardy_df['air_date'] = pd.to_datetime(jeopardy_df['AirDate']) # Assuming 'AirDate' column exists
jeopardy_df['year'] = jeopardy_df['air_date'].dt.year
plt.figure(figsize=(12, 5))
# Changing 'value' to 'Value' to match the actual column name
value_by_year = jeopardy_df.groupby('year')['Value'].mean() # Changed to 'Value'
sns.lineplot(x=value_by_year.index, y=value_by_year.values, marker='o', color='red')
plt.xlabel('Year')
plt.ylabel('Average Question Value ($)')
plt.title('Average Jeopardy Question Value Over Time')
plt.show()


#Word Cloud of Most Frequent Words**

from wordcloud import WordCloud

# Changing 'question' to 'Question' to match the actual column name
text = ' '.join(jeopardy_df['Question'].dropna()) # Changed to 'Question'
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Most Common Words in Jeopardy Questions')
plt.show()


#Heatmap of Category vs. Value**

plt.figure(figsize=(12, 6))
# Changing 'category' and 'value' to 'Category' and 'Value' respectively
category_value_avg = jeopardy_df.groupby('Category')['Value'].mean().sort_values(ascending=False).head(10) # Changed to 'Category' and 'Value'
sns.heatmap(category_value_avg.to_frame(), cmap='Blues', annot=True, fmt='.0f')
plt.xlabel('Average Value ($)')
plt.ylabel('Category')
plt.title('Top 10 Categories with Highest Average Question Value')
plt.show()

```
